# 3. Data Structures

A **data structure** is a way to organize multiple pieces of data.

In this section we will cover:
- lists
- tuples
- dictionaries

## Lists

A **list** stores multiple items in order.

Lists are:
- ordered
- changeable
- written with square brackets `[]`

````python
prices = [72.5, 73.1, 74.0, 73.8, 74.6]
products = ["WTI", "Brent", "Henry Hub"]

print(prices)
print(products)
````

### List Indexing

Similar to strings, you can access individual items inside a list using their position.

**Remember:** Python starts counting at **0**, not 1.

````python
print(prices[0])   # first item
print(prices[1])   # second item
print(products[2]) # third item
````

## Tuples

A **tuple** is like a list, except its contents are **immuatable.**

Tuples are:
- ordered
- not changeable
- written with parentheses `()`

````python
contract_months = ("Jan", "Feb", "Mar", "Apr")

print(contract_months)
print("First month:", contract_months[0])
````

## Dictionaries

A **dictionary** stores data as **key-value pairs**.

Think of it like a lookup table.

````python
hub_prices = {
    "Houston": 31.20,
    "North": 34.10,
    "South": 30.95,
    "West": 29.80
}

print(hub_prices)

# You can update dictionary values!
hub_prices["Houston"] = 31.75
hub_prices["East"] = 33.40

print("Updated dictionary:", hub_prices)
````

# 4. Control Flow (if/else statements)

A program often needs to **choose** what to do.

Python uses:
```python
if condition:
    do something
elif another_condition:
    do something else
else:
    do a fallback action
```

Python uses **indentation** to show which lines belong inside the condition.

````python
temperature = 85

if temperature > 90:
    print("Very hot day.")
elif temperature > 75:
    print("Warm day.")
else:
    print("Cool day.")
````

## Common comparison operators
- `==` equal to
- `!=` not equal to
- `>` greater than
- `<` less than
- `>=` greater than or equal to
- `<=` less than or equal to

````python
wti_price = 74.25
brent_price = 77.10

print(wti_price > 70)
print(brent_price < 75)
print(wti_price == 74.25)
print(wti_price != brent_price)
````

## Logical operators
- `and`
- `or`
- `not`

````python
price = 52
volume = 120

if price > 50 and volume > 100:
    print("Both conditions are true")

if price > 60 or volume > 100:
    print("At least one condition is true")

market_closed = False
if not market_closed:
    print("The market is open")
````

# 5. Loops

A **loop** repeats an action.

The two main loops for beginners are:
- `for` loops
- `while` loops

## `for` loops

A `for` loop repeats code for each item in a sequence.

````python
markets = ["WTI", "Brent", "Henry Hub", "ULSD"]

for market in markets:
    print(market)
````

````python
prices = [31.2, 32.5, 30.9, 33.1]

for price in prices:
    if price > 32:
        prices.remove(price)

print(prices)
````

## `while` loops

A `while` loop repeats while a condition is True.

Be careful: you must change something inside the loop, or it can run forever.

````python
countdown = 5
while countdown > 0:
    print(countdown)
    countdown = countdown - 1
print("Lift off!")
````

# 6. Functions

A **function** is a reusable block of code that performs a specific task.

Functions can take **zero or more arguments**.

Functions help you:
- avoid repeating yourself
- organize your logic
- make code easier to understand

````python
def greet(name):
    print("Hi,", name + "!")

greet("Jordan")
greet("Alex")
````

### Functions with `return`

Sometimes a function should **give a result back**.  
That is what `return` does.

````python
def calculate_spread(price_a, price_b):
    return price_a - price_b

spread = calculate_spread(77.1, 74.0)
print(spread)
````
