# 1. Variables

A **variable** is a named container for a value.

You can think of a variable as a label attached to a piece of information.

### Example idea
If Henry Hub natural gas is trading at 2.85, we can store that number in a variable.

````python
henry_hub_price = 2.85
wti_price = 78.40
trader_name = "Ray"
is_market_open = True

print(henry_hub_price)
print(wti_price)
print(trader_name)
print(is_market_open)
````

### Rules for variable names
Good Python variable names:
- use letters, numbers, and underscores
- should not start with a number
- should not contain spaces
- should be descriptive

Examples:
- `power_price`
- `daily_pnl`
- `market_region`

Less helpful examples:
- `x`
- `a1`
- `thing`

````python
# Variables can also be used in calculations

north_hub_price = 34.50
south_hub_price = 31.25

spread = north_hub_price - south_hub_price

print("North Hub Price:", north_hub_price)
print("South Hub Price:", south_hub_price)
print("Spread:", spread)
````

# 2. Data Types

A **data type** tells Python what kind of value something is.

The most common beginner data types are:
- `int` → whole numbers
- `float` → decimal numbers
- `str` → text
- `bool` → True or False

````python
trade_count = 12          # int
gas_price = 3.17          # float
hub_name = "Houston"      # str
is_peak_hour = False      # bool

print(type(trade_count))
print(type(gas_price))
print(type(hub_name))
print(type(is_peak_hour))
````

### Why data types matter
Data types affect what you can do with a value.

For example:
- numbers can be added and multiplied
- strings can be joined together
- booleans are often used in decision-making

````python
# Numbers
a = 10
b = 2.5
print(a + b)
print(a * b)

# Strings
first = "Energy"
second = "Trading"
print(first + " " + second)

# Booleans
market_open = True
print("Is the market open?", market_open)
````

## Type conversion

Sometimes data comes in as the wrong type.

For example, `"45"` is text, not a number.  
If we want to do math with it, we need to convert it.

````python
delivery_hours_text = "24"
delivery_hours_number = int(delivery_hours_text)

print(delivery_hours_text, type(delivery_hours_text))
print(delivery_hours_number, type(delivery_hours_number))

temperature_text = "98.6"
temperature_number = float(temperature_text)

print(temperature_text, type(temperature_text))
print(temperature_number, type(temperature_number))
````

# 3. Data Structures

A **data structure** is a way to organize multiple pieces of data.

In this section we will cover:
- lists
- tuples
- dictionaries
- sets

## Lists

A **list** stores multiple items in order.

Lists are:
- ordered
- changeable
- written with square brackets `[]`

````python
daily_prices = [32.10, 31.85, 33.20, 34.05, 32.90]

print(daily_prices)
print("First price:", daily_prices[0])
print("Third price:", daily_prices[2])

# Lists can be changed
daily_prices[1] = 32.00
print("Updated list:", daily_prices)

# Add a new value
daily_prices.append(33.50)
print("After append:", daily_prices)
````

### Useful list ideas
With lists, you can:
- store daily settlements
- store hourly prices
- store trade IDs
- store names of hubs or regions

## Tuples

A **tuple** is similar to a list, but it is not meant to be changed.

Tuples are:
- ordered
- not changeable
- written with parentheses `()`

````python
contract_months = ("Jan", "Feb", "Mar", "Apr")

print(contract_months)
print("First month:", contract_months[0])
````

## Dictionaries

A **dictionary** stores data as **key-value pairs**.

Think of it like a labeled lookup table.

````python
hub_prices = {
    "Houston": 31.20,
    "North": 34.10,
    "South": 30.95,
    "West": 29.80
}

print(hub_prices)
print("Houston price:", hub_prices["Houston"])

hub_prices["Houston"] = 31.75
hub_prices["East"] = 33.40

print("Updated dictionary:", hub_prices)
````

## Sets

A **set** stores unique values only.

Sets are useful when you want to remove duplicates.

````python
plants = {"Plant A", "Plant B", "Plant A", "Plant C"}

print(plants)
print("Notice that duplicates are removed automatically.")
````

# 4. Control Flow

**Control flow** means making decisions in code.

The main tool for this is the `if` statement.

### Structure
```python
if condition:
    do something
elif another_condition:
    do something else
else:
    do a fallback action
```

Python uses **indentation** to show which lines belong inside the condition.

````python
spread = 2.75

if spread > 0:
    print("The spread is positive.")
elif spread < 0:
    print("The spread is negative.")
else:
    print("The spread is exactly zero.")
````

## Comparison operators

Common comparison operators:
- `>` greater than
- `<` less than
- `>=` greater than or equal to
- `<=` less than or equal to
- `==` equal to
- `!=` not equal to

````python
load = 1850

if load >= 2000:
    print("High load")
else:
    print("Load is below the high-load threshold")
````

## Logical operators

You can combine conditions with:
- `and`
- `or`
- `not`

````python
price = 52
volume = 120

if price > 50 and volume > 100:
    print("Both conditions are true")

if price > 60 or volume > 100:
    print("At least one condition is true")

market_closed = False
if not market_closed:
    print("The market is open")
````

# 5. Loops

A **loop** repeats an action.

The two main loops for beginners are:
- `for` loops
- `while` loops

## `for` loops

A `for` loop is great when you want to go through items one by one.

````python
prices = [31.2, 32.5, 30.9, 33.1]

for price in prices:
    print("Observed price:", price)
````

````python
hubs = ["Houston", "North", "South"]

for hub in hubs:
    print("Hub:", hub)
````

## Running totals inside loops

A very common task in analytics is building a total, average, or count.

````python
daily_pnl = [1200, -300, 450, 800, -150]

total_pnl = 0

for value in daily_pnl:
    total_pnl = total_pnl + value

print("Total PnL:", total_pnl)
````

## `while` loops

A `while` loop repeats **as long as a condition is true**.

Be careful: if the condition never becomes false, the loop can run forever.

````python
hour = 1

while hour <= 5:
    print("Processing hour", hour)
    hour = hour + 1
````

### Pause and try

Then create a variable called `total_volume` and use a loop to add them together.

````python
# One possible solution

volumes = [100, 250, 400, 150]
total_volume = 0

for volume in volumes:
    print("Volume:", volume)
    total_volume = total_volume + volume

print("Total volume:", total_volume)
````

# 6. Functions

A **function** is a reusable block of code.

Functions help you:
- avoid repeating yourself
- organize your logic
- make code easier to understand

````python
def greet_student(name):
    print("Welcome to Python,", name + "!")

greet_student("Jordan")
greet_student("Alex")
````

## Functions with `return`

Sometimes a function should **give a result back**.  
That is what `return` does.

````python
def calculate_spread(price_a, price_b):
    spread = price_a - price_b
    return spread

result = calculate_spread(34.25, 31.80)
print("Spread:", result)
````

## Functions with default values

You can give parameters a default value.

````python
def calculate_trade_value(quantity, price=1.0):
    return quantity * price

print(calculate_trade_value(100))
print(calculate_trade_value(100, 72.50))
````

### Why functions matter in analytics
If you need to calculate something many times, a function saves time and reduces mistakes.

Examples:
- spread calculations
- daily PnL calculations
- average price calculations
- risk flags

### Pause and try

````python
# One possible solution

def average_of_two_numbers(a, b):
    return (a + b) / 2

print(average_of_two_numbers(10, 14))
````

# 7. Libraries

A **library** is a collection of code written by other people that you can use.

Instead of building everything from scratch, you can import tools that already exist.

In this section, we will use:
- `math`
- `statistics`
- `pandas`
- `matplotlib`

## Standard library example: `math`

````python
import math

value = 16

print("Square root:", math.sqrt(value))
print("Rounded up:", math.ceil(3.2))
print("Rounded down:", math.floor(3.8))
````

## Standard library example: `statistics`

This library is useful for basic summary statistics.

````python
import statistics

settlement_prices = [31.2, 31.8, 32.5, 30.9, 33.1]

print("Mean:", statistics.mean(settlement_prices))
print("Median:", statistics.median(settlement_prices))
print("Standard deviation:", round(statistics.stdev(settlement_prices), 4))
````

## Third-party libraries

Some libraries are not built into basic Python, but Google Colab already includes many popular ones.

Two very important libraries for data work are:
- `pandas` for tables
- `matplotlib` for charts

````python
import pandas as pd

data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "Houston_Price": [31.2, 32.0, 31.6, 33.1, 32.8],
    "North_Price": [34.1, 33.8, 34.5, 35.0, 34.7]
}

df = pd.DataFrame(data)
df
````

````python
# A quick look at summary statistics
df.describe()
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot(df["Day"], df["Houston_Price"], marker="o", label="Houston")
plt.plot(df["Day"], df["North_Price"], marker="o", label="North")
plt.title("Sample Daily Prices")
plt.xlabel("Day")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
````

# Mini Project: Bring the ideas together

Now let's combine several topics:
- variables
- lists
- loops
- functions
- libraries

We will analyze a very small set of fictional daily hub prices.

````python
prices = [29.8, 30.4, 31.1, 30.6, 32.0, 31.5]

def average_price(price_list):
    return sum(price_list) / len(price_list)

def max_price(price_list):
    return max(price_list)

def min_price(price_list):
    return min(price_list)

print("Average price:", average_price(prices))
print("Maximum price:", max_price(prices))
print("Minimum price:", min_price(prices))
````

````python
import statistics

prices = [29.8, 30.4, 31.1, 30.6, 32.0, 31.5]

avg_price = sum(prices) / len(prices)
volatility = statistics.stdev(prices)

print("Average:", round(avg_price, 2))
print("Volatility proxy (standard deviation):", round(volatility, 4))

if avg_price > 31:
    print("Average price is relatively elevated in this sample.")
else:
    print("Average price is not above 31 in this sample.")
````

````python
import pandas as pd
import matplotlib.pyplot as plt

mini_df = pd.DataFrame({
    "Day": [1, 2, 3, 4, 5, 6],
    "Price": [29.8, 30.4, 31.1, 30.6, 32.0, 31.5]
})

mini_df
````

````python
plt.figure(figsize=(8, 4))
plt.plot(mini_df["Day"], mini_df["Price"], marker="o")
plt.title("Mini Project: Daily Price Path")
plt.xlabel("Day")
plt.ylabel("Price")
plt.grid(True)
plt.show()
````
